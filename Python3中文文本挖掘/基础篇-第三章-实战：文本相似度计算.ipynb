# chapter03-实战：文本相似度计算

    有了前面的一些基础之后，我们来做一个小实战，把前面学过的内容做一个总结

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from gensim import corpora, models
from collections import Counter

import numpy as np
import os
import jieba
import logging

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s',
                    level=logging.INFO)
os.listdir('./datasets/')
```

    这里我们使用从网上获取到的一些评论作为语料集，也就是目录中的online_reviews_texts.txt

```python
data_dir = './datasets/'
text_fname = 'online_reviews_texts.txt'
textfile_path = os.path.join(data_dir, text_fname)

with open(textfile_path, encoding='utf-8') as f:
    docs = f.readlines()
print(docs[:10])
print(f'size = {len(docs)}')
```

```python
seg_list = [jieba.lcut(doc) for doc in docs]
print(seg_list[:10])
```

```python
def load_stopwords(filepath=None):
    stoplist = [
        line.strip() for line in open(filepath, encoding='utf-8').readlines()
    ]
    return list(set(stoplist))


stoplist = load_stopwords(filepath='./datasets/stopwords_list.txt')
stoplist.append('\n')
corpus = [[token for token in each if token not in stoplist]
          for each in seg_list]
corpus_like_eng = [' '.join(each) for each in corpus]
print(corpus_like_eng[:10])
```

    同样的套路，读取文本，分词，去除停用词，制作语料。需要注意的是这里用空格去join了词语，目的是为了制作类似英文写法的句子，以便调用sklearn进行向量化。

```python
tfidf_vect = TfidfVectorizer(max_df=0.95, min_df=2, max_features=400)
X_tfidf = tfidf_vect.fit_transform(corpus_like_eng)
features = tfidf_vect.get_feature_names()
print(X_tfidf.shape)
print(tfidf_vect.get_feature_names())
```

    这里我们为了方便举例，只提取400个特征词。

```python
feat_arr = X_tfidf.toarray()


def get_cosine_sim(arr1, arr2):
    """欧氏空间中的余弦相似度"""
    return np.dot(arr1, arr2) / (np.linalg.norm(arr1) * np.linalg.norm(arr2))


def get_sim_matrix(feat_mat):
    """计算文本相似度矩阵，采用余弦相似度"""
    n_size = X_tfidf.shape[0]
    sim = np.empty([n_size, n_size])
    for i in range(n_size):
        for j in range(i + 1):
            sim[i][j] = get_cosine_sim(feat_arr[i], feat_arr[j])
    return np.matrix(sim)


sim_mat = get_sim_matrix(feat_arr)
print(sim_mat)
```

    定义余弦相似度，计算文本相似度矩阵，由于有1439条文本，因而得到的矩阵就是（1439*1439），矩阵第（i，j）位置上的元素代表第（i+1）条评论和第（j+1）条评论的相似度。

```python
# 获取重心，寻找最有代表性评论
center = np.mean(feat_arr, axis=0)
center_sim = [get_cosine_sim(center, feat_arr[i]) for i in range(feat_arr.shape[0])]
idx = center_sim.index(max(center_sim)) # 与重心最相似的，认为最有代表性
print(docs[idx])
```

    有些时候我们希望找到“最有代表性”的评论，如何找呢？这里提供的思路是，先计算所有评论的“重心”，也就是将所有评论向量化后的array取算数平均（类比三角形重心的计算方法：x=（x1+x2+x3)/3, y=(y1+y2+y3)/3。然后将所有评论中，与“重心”最为相似的，认定为最有代表性评论。（仍然使用余弦相似度）
